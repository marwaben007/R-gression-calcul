```python
# put your data here
X = [6,8,10,15,20]
y = [1.8,1.3,1.1,0.7,0.4]
```

```python
import numpy as np 
from math import *
import matplotlib.pyplot as plt

X = np.array(X)
y = np.array(y)
moyx=np.mean(X)
moyy=np.mean(y)
varx=np.mean(X**2)-np.mean(X)**2
vary= np.mean(y**2)-np.mean(y)**2
ecartypex=sqrt(varx)
ecartypey=sqrt(vary)
beta1=(np.mean(X*y)-np.mean(X)*np.mean(y))/(np.mean(X**2)-np.mean(X)**2)
beta0=np.mean(y)- np.mean(X)*((np.mean(X*y)-np.mean(X)*np.mean(y))/(np.mean(X**2)-np.mean(X)**2))
rdetermination=(np.mean(X*y)-np.mean(X)*np.mean(y))/(ecartypex*ecartypey)
print("la moyenne de X :", np.mean(X))
print("la moyenne de y :",np.mean(y))
print("Sigma(x_i²) :" ,np.sum(X**2))
print("Sigma(y_i²) :",np.sum(y**2))
print("Sigma(x_i*y_i) :", np.sum(X*y))
print("moyenne(X*y) - moyenne(X)*moyenne(y)------cov(x,y) :",np.mean(X*y)-np.mean(X)*np.mean(y))
print("moyenne(X²) - (moyenne(X))²-------var(x) :", np.mean(X**2)-np.mean(X)**2)
print("moyenne(Y²) - (moyenne(Y))²-------var(y) :", np.mean(y**2)-np.mean(y)**2)
print("beta1:",(np.mean(X*y)-np.mean(X)*np.mean(y))/(np.mean(X**2)-np.mean(X)**2))
print("beta0:",np.mean(y)- np.mean(X)*((np.mean(X*y)-np.mean(X)*np.mean(y))/(np.mean(X**2)-np.mean(X)**2)))
print("r coeff de determination:",(np.mean(X*y)-np.mean(X)*np.mean(y))/(ecartypex*ecartypey))
print("r carré:",rdetermination**2)
```

```python
predection=[]
for i in range(len(X)):
    Ypred=beta1*X[i]+beta0
    predection.append(Ypred)
print("y prdeiction: ",predection)
```

```python
erreur=[]
for i in range(len(predection)):
    Resud=y[i]-predection[i]
    erreur.append(Resud)
print("Liste des residu: ",erreur)
```

```python
erreurCaree=[]
for i in range(len(erreur)):
    erc=erreur[i]**2
    erreurCaree.append(erc)
print("Liste des residu carré:",erreurCaree)
```

```python
ecartTypPredi=(1/(len(erreurCaree)-2))*sum(erreurCaree)
print("la variance prédite:",ecartTypPredi)
```

```python
EstimVarianceBeta0=(ecartTypPredi/len(X))*(1+(moyx**2/varx))
EstimVarianceBeta1=ecartTypPredi/(len(X)*varx)
print("estimation de la variance de beta0:",EstimVarianceBeta0)
print("estimation de la variance de beta1:",EstimVarianceBeta1)
print("estimation de l ecart type de beta0:",sqrt(EstimVarianceBeta0))
print("estimation de l ecart type de beta1:",sqrt(EstimVarianceBeta1))
```
